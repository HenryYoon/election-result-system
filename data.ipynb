```python
import requests
import pandas as pd 
from tqdm import tqdm
import time

api_code = pd.read_csv('D:\\coding\\선거인명부 조사\\nec_urls_1일차.csv')
# elections = ['20250603','20240410','20220601','20220309','20200415','20180613','20180509','20160413']
elections = ['20170509']

df = pd.DataFrame()

for election in elections:
    for row in tqdm(api_code.itertuples(), total=len(api_code), desc=f"{election} 선거"):
        url = 'http://apis.data.go.kr/9760000/ElcntInfoInqireService/getVtdsElcntInfoInqire'
        params ={'serviceKey' : 'iekz5HrDyh68jtoJkTI7uIBd9lDLt+oeBobNL2Iny9dzUVX7+MWHZqsMIJtyiwjSxQqvzSz5FAtjplZKCk1+zA==', 
                'pageNo' : '1', 
                'numOfRows' : '100', 
                'resultType' : 'json', 
                'sgId' : election, 
                'sdName' : row.cityName, 
                'wiwName' : row.townName}
        
        response = requests.get(url, params=params)

        try:
            data = pd.json_normalize(response.json()['response']['body']['items']['item'])
            df = pd.concat([df, data], axis=0)
        except:
            pass

        time.sleep(2)

df.to_csv('2017대선_선거인명부.csv', index=False)
```

```python
# 가장 안전하고 확실한 스크래핑 시나리오 예시
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
# 1. 일단 정상적인 경로로 페이지에 접속 (세션 및 쿠키 자동 생성됨)
driver.get("https://info.nec.go.kr/main/showDocument.xhtml?electionId=0020250603&topMenuId=VC&secondMenuId=VCAP02&cityCode=1100&townCode=1101&dateCode=1&timeCode=07")

try:
    # 4. [진짜 버튼 클릭] 드디어 대망의 검색 버튼 클릭!
    # 스크린샷에 나온 <input type="image"> 버튼을 정확히 조준해서 클릭 이벤트를 발생시킵니다.
    search_button = driver.find_element(By.CSS_SELECTOR, "#spanSubmit input[type='image']")
    search_button.click()
    
    # 5. 로딩 대기 후 데이터 파싱
    time.sleep(3)
    print("성공적으로 조회를 마쳤습니다!")
    # html = driver.page_source ... 데이터 수집 진행

except Exception as e:
    print("에러 발생:", e)

html_source = driver.page_source
html_source
# 4. 브라우저가 알아서 받아온 화면(HTML)에서 데이터 파싱하기
```

```python
from bs4 import BeautifulSoup

html = BeautifulSoup(html_source, 'html.parser')
table = html.select("#table01")

headers = [th.text.strip() for th in table[0].find('thead').find_all('th')]

# 2. 바디(데이터행) 추출하기
rows = table[0].find('tbody').find_all('tr')

data_list = []
for row in rows:
    # 행 안의 td(셀)들을 다 찾아서 텍스트만 리스트로 만듭니다.
    cells = row.find_all('td')
    row_data = [cell.text.strip() for cell in cells]
    data_list.append(row_data)

import pandas as pd

df = pd.DataFrame(data_list, columns=headers)
df
```
